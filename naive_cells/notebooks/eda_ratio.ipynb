```python
import pandas as pd
```

The project description is available [here](https://clients.adaptivebiotech.com/pub/emerson-2013-jim). Downloaded all the naive cells data (CD4+, CD8+). Unfortunately, this project doesn't provide HLA allele info.

```python
meta = pd.read_csv('../ratio/metadata.tsv', sep='\t')
meta
```

Split the data into two parts

```python
cd4_meta = meta[meta.sample_name.str.contains('CD4')]
```

```python
cd8_meta = meta[meta.sample_name.str.contains('CD8')]
```

Raw data has multiple columns. We need the amino acid sequence of the clonotype, its V gene and J gene. In addidtion to that we should filter out unfunctional sequences -- the ones which have \* or \_ signs. 

```python
example_data = pd.read_csv('../ratio/data/Healthy_Subject_4_CD4_Naive.tsv', sep='\t')
```

Let's assemble all the raw data together, filter out unfunctional sequences and select only those sequences which are unique between samples and can be assembled only using one nucleotide sequence. The resulting sets of naive cells are saved into files.

```python
datasets_to_concat = []
for sample in cd4_meta.sample_name:
    cur_data = pd.read_csv(f'../ratio/data/{sample}.tsv', sep='\t', low_memory=False)
    cur_data = cur_data[(~cur_data.amino_acid.isna()) & (cur_data.amino_acid.str.isalpha())].reset_index(drop=True)
    datasets_to_concat.append(cur_data[['sample_name', 'amino_acid', 'v_gene', 'j_gene']])
```

```python
full_cd4_data = pd.concat(datasets_to_concat)
```

```python
unique_cd4_data = full_cd4_data.groupby(by='amino_acid', as_index=False).count()
unique_cd4_data = set(unique_cd4_data[unique_cd4_data.sample_name == 1].amino_acid)
len(unique_cd4_data)
```

```python
nonunique_cd4_data = full_cd4_data.groupby(by='amino_acid', as_index=False).count()
nonunique_cd4_data = set(nonunique_cd4_data[nonunique_cd4_data.sample_name != 1].amino_acid)
len(nonunique_cd4_data)
```

```python
cd4_data = full_cd4_data[full_cd4_data.amino_acid.isin(unique_cd4_data)]
cd4_data
```

```python
cd4_data.to_csv('../ratio/naive_cd4_clonotypes.csv', index=False)
```
