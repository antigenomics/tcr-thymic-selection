Reading the raw rep-seq data. We can find the info on V/D/J segments which were used to form this clonotype, its frequency in sample, count of times it was assembled using spesific nucleotide sequence. 

```python
pd.read_csv('/projects/fmba_covid/1_data_links/hip_full/HIP00813.txt', sep='\t')
```

Reading theough all the five raw files and concatenating them.

```python
datasets_to_concat = []
for run in selected_data.file_name:
    datasets_to_concat.append(pd.read_csv('/projects/fmba_covid/1_data_links/' + run, sep='\t'))
```

```python
raw_data_concat = pd.concat(datasets_to_concat)
```

```python
raw_data_concat = raw_data_concat[raw_data_concat.cdr3aa.str.isalpha()][['cdr3aa', 'count']]
```

We want to have a look at the naive clonotypes. It means that they would be very rare (assembpled only in one sample, only once). To choose such clonotypes let's group them:
* by how many samples have this clonotype
* by how many cells we sequenced which correspond to this cdr3aa (amino acid sequence of CDR3)

As the result we would take the intersection of these two sets of clonotypes

```python
grouped_data_per_file = raw_data_concat.groupby(by=['cdr3aa'], as_index=False).count()
grouped_data_per_file.sort_values(by='count')
```

```python
unique_clones_per_file = grouped_data_per_file[grouped_data_per_file['count'] == 1]
unique_clones_per_file
```

```python
grouped_data_per_sum = raw_data_concat.groupby(by=['cdr3aa'], as_index=False).sum()
grouped_data_per_sum.sort_values(by='count')
```

We can see that there are no clonotypes which were assembpled only once, but there are many which were assempled twice (mb some effect of sequencing features). Let's assume clonotypes with count equal to 2 as naive.

```python
unique_clones_per_sum = grouped_data_per_sum[grouped_data_per_sum['count'] == 2]
unique_clones_per_sum
```

```python
len(set(unique_clones_per_file.cdr3aa).intersection(set(unique_clones_per_sum.cdr3aa)))
```

```python
unique_clones_per_sum[['cdr3aa']].to_csv('naive_clonotypes.csv', index=False)
```
