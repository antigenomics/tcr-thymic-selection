```python
naive_cd4 = pd.read_csv('./tcr-thymic-selection/naive_cells/ratio/naive_cd4_clonotypes.csv.gz')
naive_cd4
```

```python
dif = pd.Series(list(set(naive_cd4.v_gene).difference(set(olga_trb.V_gene))))
dif.index = dif
dif = dif.apply(lambda x: x if not('-1' in x) else x[:-2])
dif_dict = dif.to_dict()
dif_dict
```

```python
naive_cd4.v_gene = naive_cd4.v_gene.apply(lambda x: dif_dict[x] if x in dif_dict.keys() else x)
naive_cd4
```

```python
dif_set = set(naive_cd4.v_gene).difference(set(olga_trb.V_gene))
dif_set = dif_set.difference(set(['TRBV6-2/06-3', 'TRBV4-2/04-3', 'TRBV12-3/12-4']))
dif_set
```

```python
naive_cd4_freq = naive_cd4.v_gene.value_counts()
naive_cd4_freq = naive_cd4_freq.drop(list(dif_set))
naive_cd4_freq
```

```python
olga_trb_v_freq = olga_trb.V_gene.value_counts()#/olga_trb.V_gene.value_counts().sum()
olga_trb_v_freq
```

```python
olga_trb_v_freq = olga_trb_v_freq.append(pd.Series({'TRBV6-2/06-3': olga_trb_v_freq['TRBV6-2'] + olga_trb_v_freq['TRBV6-3']}))
olga_trb_v_freq = olga_trb_v_freq.append(pd.Series({'TRBV4-2/04-3': olga_trb_v_freq['TRBV4-2'] + olga_trb_v_freq['TRBV4-3']}))
olga_trb_v_freq = olga_trb_v_freq.append(pd.Series({'TRBV12-3/12-4': olga_trb_v_freq['TRBV12-3'] + olga_trb_v_freq['TRBV12-4']}))
olga_trb_v_freq = olga_trb_v_freq.drop(['TRBV6-2', 'TRBV6-3', 'TRBV4-2', 'TRBV4-3', 'TRBV12-3', 'TRBV12-4'])
olga_trb_v_freq
```

```python
set(olga_trb_v_freq.index).difference(set(naive_cd4_freq.index))
```

```python
set(set(naive_cd4_freq.index)).difference(set(olga_trb_v_freq.index))
```

```python
naive_cd4_freq['TRBV4-2/04-3'] = naive_cd4_freq['TRBV4-2/04-3'] + naive_cd4_freq['TRBV4-2']
naive_cd4_freq = naive_cd4_freq.drop('TRBV4-2')
naive_cd4_freq
```

```python
set(set(naive_cd4_freq.index)).difference(set(olga_trb_v_freq.index))
```

```python
dif_set = set(olga_trb_v_freq.index).difference(set(naive_cd4_freq.index))
dif_set
```

```python
olga_trb_v_freq = olga_trb_v_freq.drop(list(dif_set))
olga_trb_v_freq
```

```python
olga_trb_v_freq = olga_trb_v_freq/olga_trb_v_freq.sum()
olga_trb_v_freq
```

```python
naive_cd4_freq = naive_cd4_freq/naive_cd4_freq.sum()
```

```python
olga_trb_v_freq = olga_trb_v_freq.sort_values(ascending=False)
naive_cd4_freq = naive_cd4_freq.loc[olga_trb_v_freq.index]
```

```python
plt.figure(figsize=(45, 6))
plt.title("TRB_V_freq")
plt.xlim(-0.5, len(olga_trb_v_freq))
plt.bar(olga_trb_v_freq.index, olga_trb_v_freq, alpha = 0.5, align='center', label = 'OLGA')
plt.bar(naive_cd4_freq.index, naive_cd4_freq, alpha = 0.5, align='center', label = 'naive_cd4_ratio')
plt.legend()
plt.show()
```

```python
cd4_j_freq = naive_cd4.j_gene.value_counts()
cd4_j_freq
```

```python
olga_trb_j_freq = olga_trb.J_gene.value_counts()
olga_trb_j_freq
```

```python
set(olga_trb_j_freq.index).difference(set(cd4_j_freq.index))
```

```python
set(cd4_j_freq.index).difference(set(olga_trb_j_freq.index))
```

```python
cd4_j_freq = cd4_j_freq[olga_trb_j_freq.index]
```

```python
cd4_j_freq = cd4_j_freq/cd4_j_freq.sum()
olga_trb_j_freq = olga_trb_j_freq/olga_trb_j_freq.sum()
```

```python
plt.figure(figsize=(45, 6))
plt.title("TRB_J_freq")
plt.xlim(-0.5, len(olga_trb_j_freq))
plt.bar(olga_trb_j_freq.index, olga_trb_j_freq, alpha = 0.5, align='center', label = 'OLGA')
plt.bar(cd4_j_freq.index, cd4_j_freq, alpha = 0.5, align='center', label = 'naive_cd4_ratio')
plt.legend()
plt.show()
```
